Feature Engineering

```python
#Generate new features
#Whether native or not
df['is_native'] = df['native.country'].apply(lambda x: 1 if x == 'United-States' else 0)
```

```python
#Age to categorical variable
bins = [0, 18, 35, 50, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
#Hours per week to categorical
df['hours_per_week_bins'] = pd.cut(df['hours.per.week'], bins=[0, 20, 40, 60, 80,100], labels=['very part-time','part-time', 'full-time', 'over-time', 'extreme'])
```

```python
#simlifying capital from gain and loss to a net score
df['net_capital'] = df['capital.gain'] - df['capital.loss']
```

```python
#Work class
#Private is far and away the most prevalent work class and a class of ?
df['workclass'].value_counts()
```

```python
import numpy as np
df['workclass'] = df['workclass'].replace('?', np.nan)
```

```python
df['workclass']
```

```python
#function to replace nan with a random value from the classes present , applied when the classes are very even
def replace_nan(column, class_weights):
    non_nan_values = column.dropna().unique()
    weights = [class_weights[val] for val in non_nan_values]
    return column.apply(lambda x: np.random.choice(non_nan_values, p=weights) if pd.isna(x) else x)
```

```python
df['workclass']=replace_nan(df['workclass'],df['workclass'].value_counts(normalize=True))
```

```python
df['workclass'].value_counts()
```

```python
#Grouping the similar classes in workclass
df['workclass'] = df['workclass'].replace({
    'Federal-gov': 'Government',
    'Local-gov': 'Government',
    'State-gov': 'Government',
    'Self-emp-inc': 'Self-employed',
    'Self-emp-not-inc': 'Self-employed',
    'Without-pay': 'Unemployed',
    'Never-worked': 'Unemployed'
})
```

```python
def aggregate_education_level(level):
    if level in ['Preschool', '1st-4th', '5th-6th', '7th-8th', '9th', '10th', '11th', '12th']:
        return 'School'
    elif level in ['HS-grad']:
        return 'High School Graduate'
    elif level in ['Some-college', 'Assoc-voc', 'Assoc-acdm']:
        return 'Some College or Associate'
    elif level in ['Bachelors']:
        return 'Bachelors'
    elif level in ['Masters']:
        return 'Masters'
    else:
        return 'Doctorate or Professional'

df['education'] = df['education'].apply(aggregate_education_level)
```

```python
#marital status 
df['marital.status'].value_counts()
```

```python
def aggregate_marital_status(status):
    if status in ['Never-married']:
        return 'Single'
    elif status in ['Married-civ-spouse', 'Married-AF-spouse']:
        return 'Married'
    elif status in ['Divorced', 'Separated', 'Married-spouse-absent']:
        return 'Previously Married'
    else: # Widowed
        return 'Widowed'

df['marital.status'] = df['marital.status'].apply(aggregate_marital_status)
```

```python
df['occupation']=replace_nan(df['occupation'],df['occupation'].value_counts(normalize=True))
```

```python
def aggregate_occupation(job):
    if job in ["Machine-op-inspct", "Transport-moving", "Handlers-cleaners", "Farming-fishing", "Craft-repair"]:
        return 'Blue Collar'
    elif job in ["Prof-specialty", "Exec-managerial", "Adm-clerical", "Sales", "Tech-support" ]:
        return 'White Collar'
    elif job in ["Protective-serv" , "Priv-house-serv"]:
        return 'Service'
    else:
        return job

df['occupation'] = df['occupation'].apply(aggregate_occupation)
```

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
```

Comparing all the different transformations to find the best one

```python
def logTrans(feature):  
    
    logTr = ColumnTransformer(transformers=[("lg", FunctionTransformer(np.log1p), [feature])])

    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1)
    plt.title("Distribution before Transformation", fontsize=15)
    sns.histplot(df[feature], kde=True, color="red")
    plt.subplot(1,2,2)
    
    data_log = pd.DataFrame(logTr.fit_transform(df))
    plt.title("Distribution after Transformation", fontsize=15)
    sns.histplot(data_log,bins=20, kde=True , legend=False)
    plt.xlabel(feature)
    plt.show()
    print(data_log.skew())
    print(data_log.kurt())
    
for feature in skewed_data:
    logTrans(feature)
```

```python
def squareTrans(feature):   
    
    logTr = ColumnTransformer(transformers=[("lg", FunctionTransformer(np.square), [feature])])

    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1)
    plt.title("Distribution before Transformation", fontsize=15)
    sns.histplot(df[feature], kde=True, color="red")
    plt.subplot(1,2,2)
    
    data_square = pd.DataFrame(logTr.fit_transform(df))
    plt.title("Distribution after Transformation", fontsize=15)
    sns.histplot(data_square,bins=20, kde=True , legend=False)
    plt.xlabel(feature)
    plt.show()
    print(data_square.skew())
    print(data_square.kurt())

    
for feature in skewed_data:
    squareTrans(feature)
```

```python
def cubeTrans(feature):   
    
    logTr = ColumnTransformer(transformers=[("lg", FunctionTransformer(lambda x: x**3), [feature])])

    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1)
    plt.title("Distribution before Transformation", fontsize=15)
    sns.histplot(df[feature], kde=True, color="red")
    plt.subplot(1,2,2)
    
    data_cube = pd.DataFrame(logTr.fit_transform(df))
    plt.title("Distribution after Transformation", fontsize=15)
    sns.histplot(data_cube,bins=20, kde=True , legend=False)
    plt.xlabel(feature)
    plt.show()
    print(data_cube.skew())
    print(data_cube.kurt())
    
for feature in skewed_data:
    cubeTrans(feature)
```

```python
def sqrtTrans(feature):   
    
    logTr = ColumnTransformer(transformers=[("lg", FunctionTransformer(lambda x: x**1/2), [feature])])

    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1)
    plt.title("Distribution before Transformation", fontsize=15)
    sns.histplot(df[feature], kde=True, color="red")
    plt.subplot(1,2,2)
    
    data_sqrt = pd.DataFrame(logTr.fit_transform(df))
    plt.title("Distribution after Transformation", fontsize=15)
    sns.histplot(data_sqrt,bins=20, kde=True , legend=False)
    plt.xlabel(feature)
    plt.show()
    print(data_sqrt.skew())
    print(data_sqrt.kurt())
    
for feature in skewed_data:
    sqrtTrans(feature)
```

```python
from sklearn.preprocessing import PowerTransformer

def yeojohnTrans(feature):   
    
    yeojohnTr = PowerTransformer(standardize=True)  

    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1)
    plt.title("Distribution before Transformation", fontsize=15)
    sns.histplot(df[feature], kde=True, color="red")
    plt.subplot(1,2,2)
    
    data_yeojohn = pd.DataFrame(yeojohnTr.fit_transform(df[feature].values.reshape(-1,1)))
    plt.title("Distribution after Transformation", fontsize=15)
    sns.histplot(data_yeojohn,bins=20, kde=True , legend=False)
    plt.xlabel(feature)
    plt.show()
    print(data_yeojohn.skew())
    print(data_yeojohn.kurt())
    
for feature in skewed_data:
    yeojohnTrans(feature)
```
