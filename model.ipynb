```python
import pandas as pd
df=pd.read_csv('new_df.csv')
```

```python
#One Hot Encoding 
from sklearn.preprocessing import OneHotEncoder
categorical_cols=['age_group','hours_per_week_bins','workclass','education','marital.status','occupation','relationship','race','sex']
encoder = OneHotEncoder(handle_unknown='ignore')
encoder.fit(df[categorical_cols])
encoded_df = pd.DataFrame(encoder.transform(df[categorical_cols]).toarray(), columns=encoder.get_feature_names_out(categorical_cols))
final_df = pd.concat([df.drop(categorical_cols, axis=1), encoded_df], axis=1)
```

```python
final_df
```

```python
from joblib import dump
with open("one_hot_encoder.pkl", "wb") as f:
    dump(encoder, f)
```

```python
final_df
```

```python
final_df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
y=final_df['income']
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
final_df['income']=y_encoded
```

```python
final_df
```

```python
df['income'].value_counts()

#Quite an imbalanced dataset
```

```python
X = final_df.iloc[:, [*range(0, 4), *range(5, final_df.shape[1])]]
```

```python
X
```

```python
y=final_df['income']
y
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
smote=SMOTE(sampling_strategy='minority')
X_sm,y_sm=smote.fit_resample(X,y.ravel())
```

```python
X_sm
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_sm=ss.fit_transform(X_sm.values)
```

```python
with open("standard_scaler.pkl", "wb") as f:
    dump(ss, f)
```

```python
X_sm
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_sm,y_sm,test_size=0.2,random_state=0,stratify=y_sm)
```

```python
y_train.shape
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score

lr=LogisticRegression()
lr.fit(X_train,y_train)
predictions=lr.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

print("\nClassification Report:")
print(classification_report(y_test, predictions))

accuracies=cross_val_score(estimator=lr,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.svm import SVC

svc = SVC(kernel='rbf')
svc.fit(X_train, y_train)
predictions = svc.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

print("\nClassification Report:")
print(classification_report(y_test, predictions))

accuracies=cross_val_score(estimator=svc,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)  
rf.fit(X_train, y_train)
predictions = rf.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

print("\nClassification Report:")
print(classification_report(y_test, predictions))

accuracies=cross_val_score(estimator=rf,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
import xgboost as xgb

xg = xgb.XGBClassifier()
xg.fit(X_train, y_train)
predictions = xg.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

print("\nClassification Report:")
print(classification_report(y_test, predictions))

accuracies=cross_val_score(estimator=xg,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))

print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score

rf = RandomForestClassifier(max_depth=20 ,min_samples_leaf= 1 ,min_samples_split=2, n_estimators=300)
rf.fit(X_train, y_train)
predictions = rf.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

print("\nClassification Report:")
print(classification_report(y_test, predictions))

accuracies=cross_val_score(estimator=rf,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
